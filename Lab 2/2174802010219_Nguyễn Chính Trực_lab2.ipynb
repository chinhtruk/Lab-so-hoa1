```python
## cho 1 bản np có 150 dòng 5 cột
import numpy as np
from sklearn.model_selection import train_test_split

# 1. Tạo mảng NumPy 150 x 5
np.random.seed(42)  # Đặt seed để kết quả nhất quán
data = np.random.rand(150, 5) * 100  # Tạo mảng ngẫu nhiên, nhân với 100 để tăng giá trị

# 2. Chia cột đầu tiên (chiều cao) thành x và cột cuối cùng (GPA) thành y
x = data[:, 0]  # Cột đầu tiên
y = data[:, -1]  # Cột cuối cùng

# 3. Chia dữ liệu x và y thành train/test (70:30)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# 4. Tạo tập dữ liệu không chồng chéo (loại bỏ giá trị trùng lặp trong x_train)
x_train_unique = np.unique(x_train)


```

```python
#Bài tập tết
import numpy as np

# Khởi tạo ma trận 3x3 với giá trị mặc định là 99
matrix = np.full((3, 3), 99, dtype=object)

# Biến theo dõi lượt chơi (0 là 'O', 1 là 'X')
turn = 0  

while True:
    # Hiển thị ma trận hiện tại
    print("\nMa trận hiện tại:")
    for row in matrix:
        print(row)

    # Nhập vị trí từ người chơi
    try:
        row, col = map(int, input(f"Người chơi {'X' if turn == 1 else 'O'} nhập vị trí (dòng, cột) [0-2 0-2]: ").split())

        # Kiểm tra nếu nhập (0,0) thì trò chơi kết thúc
        if (row, col) == (0, 0):
            print("Bạn đã nhập (0,0). Trò chơi kết thúc!")
            break

        # Kiểm tra nếu nhập sai vị trí (ngoài phạm vi ma trận)
        if row not in [0, 1, 2] or col not in [0, 1, 2]:
            print("Vị trí không hợp lệ! Vui lòng nhập lại.")
            continue

        # Kiểm tra nếu ô đã có giá trị (không phải 99)
        if matrix[row, col] != 99:
            print("Ô này đã có người chọn! Vui lòng chọn ô khác.")
            continue

        # Điền vào ma trận theo lượt chơi
        matrix[row, col] = "X" if turn == 1 else "O"

        # Kiểm tra điều kiện thắng (ba ô liền kề giống nhau)
        for i in range(3):
            # Kiểm tra hàng, cột
            if (matrix[i, 0] == matrix[i, 1] == matrix[i, 2] != 99 or 
                matrix[0, i] == matrix[1, i] == matrix[2, i] != 99):
                print(f"Người chơi {'X' if turn == 1 else 'O'} thắng!")
                exit()

        # Kiểm tra hai đường chéo
        if (matrix[0, 0] == matrix[1, 1] == matrix[2, 2] != 99 or
            matrix[0, 2] == matrix[1, 1] == matrix[2, 0] != 99):
            print(f"Người chơi {'X' if turn == 1 else 'O'} thắng!")
            exit()

        # Chuyển lượt chơi
        turn = 1 - turn  

    except ValueError:
        print("Vui lòng nhập đúng định dạng số nguyên.")
    except IndexError:
        print("Chỉ số ngoài phạm vi! Nhập giá trị từ 0 đến 2.")

print("Trò chơi kết thúc!")
```

## **Phần 1: NumPy cơ bản**

### **Bài tập 1: Tạo mảng và thao tác cơ bản**

1 Tạo một mảng NumPy với các giá trị từ 1 đến 20.\
2 Tìm tổng, giá trị lớn nhất, nhỏ nhất và trung bình của mảng.\
3 Tạo một mảng 2D (3x5) chứa các số ngẫu nhiên từ 0 đến 100.\
4 Lấy hàng thứ 2 và cột thứ 3 của mảng 2D.

```python
import numpy as np

# 1. Tạo một mảng NumPy với các giá trị từ 1 đến 20
array_1 = np.arange(1, 21)
print("Mảng từ 1 đến 20:", array_1)

# 2. Tìm tổng, giá trị lớn nhất, nhỏ nhất và trung bình của mảng
print("Tổng:", np.sum(array_1))
print("Giá trị lớn nhất:", np.max(array_1))
print("Giá trị nhỏ nhất:", np.min(array_1))
print("Trung bình:", np.mean(array_1))

# 3. Tạo một mảng 2D (3x5) chứa các số ngẫu nhiên từ 0 đến 100
array_2d = np.random.randint(0, 101, (3, 5))
print("Mảng 2D:\n", array_2d)

# 4. Lấy hàng thứ 2 và cột thứ 3 của mảng 2D
print("Hàng thứ 2:", array_2d[1])  # Hàng thứ 2 (chỉ số 1)
print("Cột thứ 3:", array_2d[:, 2])  # Cột thứ 3 (chỉ số 2)
```

### **Bài tập 2: Các thao tác nâng cao**

1 Tạo một mảng NumPy chứa 20 giá trị ngẫu nhiên từ 0 đến 1.\
2 Chuẩn hóa mảng này (đưa các giá trị về khoảng [0, 1]).\
3 Tính tích vô hướng (dot product) của hai mảng 1D: [1, 2, 3] và [4, 5,6].\
4 Tạo một ma trận 5x5 và tính định thức (determinant) và nghịch đảo của ma trận.

```python
import numpy as np

# 1. Tạo một mảng NumPy chứa 20 giá trị ngẫu nhiên từ 0 đến 1
random_array = np.random.rand(20)
print("Mảng ngẫu nhiên từ 0 đến 1:", random_array)

# 2. Chuẩn hóa mảng này (đưa các giá trị về khoảng [0, 1])
min_val = np.min(random_array)
max_val = np.max(random_array)
normalized_array = (random_array - min_val) / (max_val - min_val)
print("Mảng sau khi chuẩn hóa:", normalized_array)

# 3. Tính tích vô hướng (dot product) của hai mảng 1D
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot_product = np.dot(a, b)
print("Tích vô hướng của a và b:", dot_product)

# 4. Tạo một ma trận 5x5 và tính định thức, nghịch đảo
matrix = np.random.randint(1, 10, (5, 5))  # Tạo ma trận 5x5 với số nguyên từ 1 đến 9
print("Ma trận:\n", matrix)

# Tính định thức
determinant = np.linalg.det(matrix)
print("Định thức của ma trận:", determinant)

# Kiểm tra nếu ma trận khả nghịch (định thức khác 0) thì tính nghịch đảo
if determinant != 0:
    inverse_matrix = np.linalg.inv(matrix)
    print("Ma trận nghịch đảo:\n", inverse_matrix)
else:
    print("Ma trận không khả nghịch (định thức = 0).")
```

## **Phần 2: Pandas cơ bản**

### **Bài tập 3: Làm quen với DataFrame**
1 Tạo một DataFrame chứa thông tin sau:
| Name    | Age | Score |
|---------|-----|-------|
| Alice   | 23  | 85    |
| Bob     | 25  | 90    |
| Charlie | 22  | 78    |
| David   | 24  | 92    |
| Eva     | 21  | 88    |

2 Tính giá trị trung bình của cột "Score".\
3 Lọc các hàng có "Score" lớn hơn 85.

```python
import pandas as pd

# 1. Tạo DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [23, 25, 22, 24, 21],
    "Score": [85, 90, 78, 92, 88]
}

df = pd.DataFrame(data)
print("DataFrame:\n", df)

# 2. Tính giá trị trung bình của cột "Score"
average_score = df["Score"].mean()
print("Giá trị trung bình của Score:", average_score)

# 3. Lọc các hàng có "Score" lớn hơn 85
filtered_df = df[df["Score"] > 85]
print("Các hàng có Score > 85:\n", filtered_df)
```

## **Phần 3: Làm sạch dữ liệu**

### **Bài tập 5: Xử lý dữ liệu thiếu**
1 Tạo một DataFrame chứa các giá trị sau:

| Name    | Age | City     | Salary |
|---------|-----|----------|--------|
| Alice   | 23  | New York | 60000  |
| Bob     | NaN | Boston   | 52000  |
| Charlie | 25  | NaN      | NaN    |
| David   | 24  | Chicago  | 58000  |
| Eva     | 22  | Boston   | NaN    |

2 Điền giá trị thiếu trong cột Age bằng giá trị trung bình.\
3 Xóa các hàng có nhiều hơn 1 giá trị thiếu.\
4 Điền giá trị thiếu trong cột Salary bằng 50000.

```python
import pandas as pd

# 1. Tạo DataFrame chứa dữ liệu thiếu
data_with_missing = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [23, None, 25, 24, 22],
    "City": ["New York", "Boston", None, "Chicago", "Boston"],
    "Salary": [60000, 52000, None, 58000, None]
}

df_missing = pd.DataFrame(data_with_missing)
print("Dữ liệu ban đầu:\n", df_missing)

# 2. Điền giá trị thiếu trong cột Age bằng giá trị trung bình
df_missing["Age"].fillna(df_missing["Age"].mean(), inplace=True)
print("\nDữ liệu sau khi điền Age bằng giá trị trung bình:\n", df_missing)

# 3. Xóa các hàng có nhiều hơn 1 giá trị thiếu
df_missing = df_missing.dropna(thresh=df_missing.shape[1] - 1)  # Giữ lại hàng có ít nhất 3 giá trị không thiếu
print("\nDữ liệu sau khi xóa các hàng có nhiều hơn 1 giá trị thiếu:\n", df_missing)

# 4. Điền giá trị thiếu trong cột Salary bằng 50000
df_missing["Salary"].fillna(50000, inplace=True)
print("\nDữ liệu sau khi điền Salary bằng 50000:\n", df_missing)
```

## **Phần 4: Trực quan hóa dữ liệu với Matplotlib**

### **Bài tập 6: Biểu đồ cơ bản**

1 Tạo một biểu đồ đường biểu diễn hàm số y = x ** 2 trên khoảng [-10,10] \
2 Vẽ biểu đồ cột thể hiện điểm số (Score) của các sinh viên từ Bài tập 3.\
3 Tạo một biểu đồ tròn (pie chart) thể hiện phần trăm mỗi loại hoa trong tập dữ liệu Iris.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Biểu đồ đường biểu diễn hàm số y = x^2
x = np.linspace(-10, 10, 100)  # Tạo khoảng giá trị từ -10 đến 10
y = x ** 2  # Hàm số y = x^2

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="y = x^2", color="red", linewidth=2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Biểu đồ đường của y = x^2")
plt.legend()
plt.grid()
plt.show()

# 2. Biểu đồ cột thể hiện điểm số sinh viên
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Score": [85, 90, 78, 92, 88]
}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
plt.bar(df["Name"], df["Score"], color="blue", alpha=0.7)

# Thêm nhãn và tiêu đề
plt.xlabel("Sinh viên")
plt.ylabel("Điểm số")
plt.title("Biểu đồ cột điểm số của sinh viên")
plt.ylim(70, 100)  # Giới hạn trục y để dễ nhìn hơn
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Hiển thị biểu đồ
plt.show()

# 3. Biểu đồ tròn thể hiện phần trăm mỗi loại hoa trong tập dữ liệu Iris
file_path = "/Volumes/Data/Uni Lessions/Số hoá và quản trị thông tin số/Lab 1/Iris Dataset.csv"
iris_df = pd.read_csv(file_path)

# Đếm số lượng mỗi loại hoa
species_counts = iris_df["species"].value_counts()

# Vẽ biểu đồ tròn
plt.figure(figsize=(7, 7))
plt.pie(species_counts, labels=species_counts.index, autopct="%1.1f%%", startangle=140)
plt.title("Tỷ lệ phần trăm mỗi loại hoa trong tập dữ liệu Iris")
plt.show()
```

### **Bài tập 7: Biểu đồ nâng cao**

1 Vẽ biểu đồ phân tán (scatter plot) giữa sepal_length và sepal_width của tập dữ liệu Iris. Dùng màu sắc để phân biệt các loại hoa (species).\
2 Thêm tiêu đề, nhãn trục và chú thích cho biểu đồ.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu từ file CSV
file_path = "/Volumes/Data/Uni Lessions/Số hoá và quản trị thông tin số/Lab 1/Iris Dataset.csv"
iris_df = pd.read_csv(file_path)

# 1. Biểu đồ phân tán (scatter plot) giữa sepal_length và sepal_width của tập dữ liệu Iris
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x="sepal_length", y="sepal_width", hue="species", palette="viridis")

# 2. Thêm tiêu đề, nhãn trục và chú thích
plt.title("Biểu đồ phân tán giữa Sepal Length và Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Loại hoa")
plt.grid(True, linestyle="--", alpha=0.6)

# Hiển thị biểu đồ
plt.show()
```
